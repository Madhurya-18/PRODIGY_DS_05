```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# To read the csv file(dataset)
df = pd.read_csv('RTA Dataset.csv')
df
```

```python
# Checking the number of rows and columns in our dataset
df.shape
```

```python
# Getting basic information about the dataset
df.info()
```

```python
# Getting some statistical information of our data
df.describe()
```

```python
#count the number of missing (null) values in each column
df.isnull().sum()
```

```python
#Handling the missing values by replacing '?' with NaN
df.replace('?', np.nan, inplace=True)

# Replace NaN with 0
df.fillna(0, inplace=True)
df.isnull().sum()
```

```python
df['Accident_severity'].value_counts()
```

```python
# To visualize distribution of accident severity
sns.countplot(x=df['Accident_severity'], palette = 'Set2')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.title('Distribution of Accident severity',fontweight='bold')
```

```python
# Print column names to check for available columns
print(df.columns)

# Extract relevant columns for analysis
relevant_columns = ['Time', 'Day_of_week', 'Weather_conditions', 'Road_surface_conditions']
df_subset = df[relevant_columns]
```

```python
# Display the subset of data
print(df_subset)
```

```python
# Print column names of df_subset
print(df_subset.columns)
```

**IDENTIFYING PATTERN RELATED TO ROAD CONDITIONS**

```python
# Count accidents by road conditions
road_condition_counts = df_subset['Road_surface_conditions'].value_counts()

# Display the counts
print(road_condition_counts)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
colors = plt.cm.viridis(np.linspace(0, 1, len(road_condition_counts)))
plt.barh(road_condition_counts.index, road_condition_counts.values, color=colors, edgecolor='black', linewidth=1.2)
plt.title('Accidents by Road Conditions', fontsize=12, fontweight='bold')
plt.xlabel('Number of Accidents', fontsize=12)
plt.ylabel('Road Conditions', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(road_condition_counts):
    plt.text(value, index, str(value), color='black', fontsize=10, va='center')
plt.gca().set_facecolor('#f9f9f9')
plt.show()
```

**IDENTIFYING PATTERN RELATED TO WEATHER CONDITIONS**

```python
# Count accidents by weather conditions
weather_condition_counts = df_subset['Weather_conditions'].value_counts()

# Display the counts
print(weather_condition_counts)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
weather_condition_counts.plot(kind='line', color='darkorchid', marker='o', linestyle='--', linewidth=2, markersize=8)
plt.title('Accidents by Weather Conditions', fontsize=12, fontweight='bold')
plt.xlabel('Weather Conditions', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

for x, y in enumerate(weather_condition_counts):
    plt.text(x, y, str(y), color='black', fontsize=10, ha='center')
plt.gca().set_facecolor('#f0f0f0')
plt.legend(['Weather Conditions'], loc='best')
plt.show()
```


**IDENTIFYING PATTERNS RELATED TO TIME OF THE DAY**

```python
# Count accidents by time
time_counts = df_subset['Time'].apply(lambda x: x.split(':')[0]).value_counts().sort_index()

# Display the counts
print(time_counts)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 3)
time_counts.plot(kind='bar', color='pink', edgecolor='black')
plt.title('Accidents by Time of the Day', fontsize=14, fontweight='bold')
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Road_surface_conditions_encoded'] = le.fit_transform(df['Road_surface_conditions'])
df['Weather_conditions_encoded'] = le.fit_transform(df['Weather_conditions'])
df['Time_encoded'] = le.fit_transform(df['Time'])
```

**VISUALIZING ACCIDENT SPOTS**

```python
# Accident spots based on 'Area_accident_occured'
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Area_accident_occured', hue='Accident_severity',palette='Set2')
plt.title('Accidents by Area and Severity',fontweight='bold')
plt.xlabel('Area Where Accident Occurred')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Accident Severity')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
correlation_matrix = df[['Road_surface_conditions_encoded', 'Weather_conditions_encoded', 'Time_encoded']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Set2')
plt.title('Correlation Matrix of Contributing Factors',fontweight='bold')
plt.show()
```
